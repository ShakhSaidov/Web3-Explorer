```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

table = pd.read_csv("dataCollection_TEST.csv")
df = table[['ShareScore','BalanceScore','Label']].copy()
labels = ['DEAD', 'PUMP']

sns.relplot(data=df, x='ShareScore', y='BalanceScore', hue='Label', hue_order=labels, aspect=1.61)
plt.show()

```
