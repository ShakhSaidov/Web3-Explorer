```python
import requests
import json
from csv import writer
import pandas as pd

#GLOBAL VARIABLES
uniswapV2 = "0x7a250d5630b4cf539739df2c5dacb4c659f2488d"
uniswapV3 = "0x68b3465833fb72a70ecdf485e0e4c7bd8665fc45"

#CHANGE VARIABLES IN THE QUERY: PAIR AND TIMESTAMP
query = """
        {
            swaps(
                where: {
                    pair: "0x3797e0cd35ef72c7a8a62a378f6112823a553430",
                    timestamp_lt: 1674416460,
                }
                orderDirection: asc
                orderBy: timestamp
                first: 1000
            ) {
                sender
                to
            }
        }
    """

resp = requests.post('https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2',
                        json={'query': query})

json_data = json.loads(resp.text)
json_data = json_data["data"]["swaps"]

for eachData in json_data:
    if((eachData["sender"] == uniswapV2 or eachData["sender"] == uniswapV3) and (eachData["to"] != uniswapV2 or eachData["to"] != uniswapV3)):
        wallet = eachData["to"]
        if (wallet == uniswapV2 or wallet == uniswapV3): continue
        with open('donkeyAGAIN4.csv', 'a') as fd:
            fd.write(f"{wallet}\n")
```

```python
# ALCHEMY
import requests
import json

demourl = "https://eth-mainnet.g.alchemy.com/v2/docs-demo"
url = "https://eth-mainnet.g.alchemy.com/v2/RFZwgvbVeEzxo2bZwX4sR6AG1p0TTaso"
uniswapV2 = "0x7a250d5630b4cf539739df2c5dacb4c659f2488d"
uniswapV3 = "0x68b3465833fb72a70ecdf485e0e4c7bd8665fc45"
contract = "0xf89e98be8fabcfd2422562a6a04593ea29e628c1"

payload = {
    "id": 1,
    "jsonrpc": "2.0",
    "method": "alchemy_getAssetTransfers",
    "params": [
        {
            "fromBlock": "0x0",
            "toBlock": "0xFB3A71",
            "contractAddresses": ["0xf89e98be8fabcfd2422562a6a04593ea29e628c1"],
            "category": ["erc20"],
            "withMetadata": True,
            "excludeZeroValue": True,
            "maxCount": "0x3e8",
            "order": "desc"
        }
    ]
}
headers = {
    "accept": "application/json",
    "content-type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)
response = response.text

responseJSON = json.loads(response)
data = responseJSON["result"]["transfers"]

wallets = []

for each in data:
    if(each["from"] == uniswapV2 or each['from'] == uniswapV3 or each["from"] == contract):
        wallets.append(each["to"])

    elif(each["to"] == uniswapV2 or each['to'] == uniswapV3 or each["to"] == contract):
       wallets.append(each["from"])

with open('donkeyAGAIN3.csv', 'a') as fd:
    for wallet in wallets:
        fd.write(f"{wallet}\n")

```
