```python
#DONKEY
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.5f' % x)

donkey = pd.read_csv("donkeyAGAIN3.csv")
holders = pd.read_csv("donkeyAGAIN4.csv")
#holders = holders[holders['Balance'] > 52000000]

holders3 = pd.merge(donkey, holders, on='HolderAddress', how='inner')
holders3

holders3 = holders3.groupby(['HolderAddress']).size().reset_index(name='counts')
holders3.sort_values('counts', ascending=False)
#holders3.to_csv("donkeyTEST.csv")
```

```python
#THANOS
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.5f' % x)

donkey = pd.read_csv("THANOS.csv")
aicapHolders = pd.read_csv("THANOSholders.csv")
aicapHolders = aicapHolders[aicapHolders['Balance'] >= 1000]

holders4 = pd.merge(donkey, aicapHolders, on='HolderAddress', how='inner')
holders4

holders4 = holders4.groupby(['HolderAddress']).size().reset_index(name='counts')
holders4.sort_values('counts', ascending=False)
```
